```python
from random import *
from tkinter import * 
from pandas import *



# Đọc file Excel chứa danh sách từ vựng
def read_excel_file(filename):
    df = read_excel(filename)
    vocab_list = df.values.tolist()
    global total_count
    total_count = len(df)
    return vocab_list

# Cập nhật câu hỏi mới khi người dùng trả lời đúng
def update_question():
    global current_word, vocab_list
    current_word = choice(vocab_list)
    vocab_label.config(text=current_word[0])
    pronunciation_label.config(text=current_word[1])
    feedback_label.config(text="")


correct_count = 0
def check_answer():
    global current_word_index, current_word, vocab_list, answered_list, correct_count
    answer = answer_entry.get().strip().lower()
    for i in current_word[2].split("|"):
        if answer != i.strip().lower():
            result = False
            continue
        else:
            result = True
            break
            
    if result == True:
        feedback_sublabel.config(text = "")
        answer_entry.delete(0, END)

                   
        # Bỏ câu hỏi hiện tại khỏi danh sách câu hỏi

        vocab_list.pop(vocab_list.index(current_word))
            
        # Nếu danh sách câu hỏi đã trả lời hết thì hiển thị thông báo
        if len(vocab_list) == 0:
            vocab_label.config(text="Bạn đã hoàn thành tất cả câu hỏi!")
            pronunciation_label.config(text="")
            answer_entry.config(state=DISABLED)
            feedback_label.config(text="Bravo!", fg = "green",font=("Helvetica", 18),image =photo_bravo,compound = LEFT)
            feedback_sublabel.config(text = "")
            correct_count += 1
            correct_label.config(text=f"Tiến độ: {correct_count} / {total_count}")
            
        else:
            correct_count += 1
            correct_label.config(text=f"Tiến độ: {correct_count} / {total_count}")
            update_question()
        
    else:
        feedback_label.config(text="Đáp án sai. Vui lòng thử lại")
        feedback_sublabel.config(text = f' Đáp án đúng: {current_word[2]}')
        
# Tạo giao diện người dùng
root = Tk()
root.title("Flashcard - Học tiếng Anh")
root.geometry("400x400")

# Đọc file Excel chứa danh sách từ vựng
vocab_list = read_excel_file("flashcard.xlsx")


# Tạo các label frame
info_frame = LabelFrame(root, text="Info")
question_frame = LabelFrame(root, text="Question")
answer_frame = LabelFrame(root, text="Answer")
feedback_frame = LabelFrame(root, text="Feedback")

# Định vị các label frame
info_frame.pack(fill="both", expand=True, padx=20, pady=10)
question_frame.pack(fill="both", expand=True, padx=20, pady=10)
answer_frame.pack(fill="both", expand=True, padx=20, pady=10)
feedback_frame.pack(fill="both", expand=True, padx=20, pady=10)

# Trang trí icon
photo_info = PhotoImage(file =r'info.png')
info_label = Label(info_frame, image=photo_info)
info_label.pack(side=LEFT,anchor='n')

photo_question = PhotoImage(file =r'question.png')
question_label = Label(question_frame, image=photo_question)
question_label.pack(side=LEFT,anchor='n')

photo_answer = PhotoImage(file =r'answer.png')
answer_label = Label(answer_frame, image= photo_answer)
answer_label.pack(side=LEFT,anchor='n')

photo_feedback = PhotoImage(file =r'feedback.png')
feedback_label = Label(feedback_frame, image=photo_feedback)
feedback_label.pack(side=LEFT,anchor='n')

photo_bravo = PhotoImage(file =r'clapping.png')

myfont = ('Helvetica', 12)
# Tạo các widget
correct_label = Label(master=info_frame, text=f"Tiến độ: {correct_count} / {total_count}", font=myfont, fg = "green")
vocab_label = Label(master=question_frame, text="", font=("Helvetica", 16))
pronunciation_label = Label(master=question_frame, text="", font=myfont, fg = "blue")
answer_entry = Entry(master=answer_frame, font=("Helvetica", 16))
answer_entry.bind("<Return>", lambda event: check_answer())
feedback_label = Label(master=feedback_frame, text="", font=myfont, fg = "red")
feedback_sublabel = Label(master=feedback_frame, text="", font=myfont)

# Định vị các widget trong các label frame
correct_label.pack(ipadx = 20, ipady=20)
vocab_label.pack(ipadx = 10, ipady=10)
pronunciation_label.pack()
answer_entry.pack(ipadx = 5, ipady=5)
feedback_label.pack(ipadx = 5, ipady=5)
feedback_sublabel.pack()


# Cập nhật câu hỏi
update_question()


root.mainloop()
```
